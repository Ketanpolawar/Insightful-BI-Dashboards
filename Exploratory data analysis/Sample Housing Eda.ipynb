STEP3:GETTING THE VISUAL ANALYSIS AND DRAWING THE CONCLUSIONS


```python
num_cols=['YearBuilt','TotalBsmtSF','GrLivArea','SalePrice']
for i in range(0,len(num_cols),2):
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.distplot(Dfa[num_cols[i]],kde=False,color='lightgreen')
    plt.subplot(122)
    sns.distplot(Dfa[num_cols[i+1]],kde=False,color='lightblue')
    plt.tight_layout()
    plt.show()
    #histogram shows the distribution of data.
    #The intervals are referred to as 'bins' and they are all the same width
```

```python
num_cols=['YearBuilt','TotalBsmtSF','GrLivArea','SalePrice']
for i in range(0,len(num_cols),2):
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    s1=sns.distplot(Dfa[num_cols[i]],kde=True,color='lightpink')
    plt.subplot(122)
    s2=sns.distplot(Dfa[num_cols[i+1]],kde=True,color='grey')
    plt.show()
    #histogram shows the distribution of data.
    #The intervals are referred to as 'bins' and they are all the same width
```

Observation:

```python
sns.set(style="whitegrid")
f,ax=plt.subplots(figsize=(15,10))
sns.violinplot(x="SaleCondition",y="SalePrice",data=Dfa,inner="box",palette="Set3",cut=2,linewidth=3,ax=ax)
sns.despine(left=True)
ax.set_title('Salecondition vs salesPrice')
ax.set_xlabel("Salecondition",alpha=0.7)
ax.set_ylabel("salePrice",alpha=0.7)
#It is similar to a box plot, with the addition of a rotated kernel density plot on each side
```

```python
#A radar chart (or spider plot, or polar chart) allows to visualize one or more series of values over multiple quantitative variables.

import plotly.express as px
import pandas as pd
food_data=pd.DataFrame(dict(item_rating=[1,5,2,4,5],criteria=['Packaging Cost','Food Quality','Ingredients Cost','Market Demand','Customer Rating']))
food_data.head()
```

```python
import plotly.express as px
import pandas as pd
food_data=pd.DataFrame(dict(item_rating=[1,5,2,4,5],criteria=['Packaging Cost','Food Quality','Ingredients Cost','Market Demand','Customer Rating']))
food_data.head()
fig=px.line_polar(food_data,r='item_rating',theta='criteria',line_close=True)
fig.update_traces(fill='toself')
fig.show()
```
